Exercise 1: Prime Numbers

```python
def is_prime(n):
    # A number less than or equal to 1 is not prime
    if n <= 1:
        return False

    # Check divisibility from 2 up to the square root of n
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False

    return True

# Input from the user
number = int(input("Enter a number: "))

# Check and display whether the number is prime
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

Exercise 2: Product of Random Numbers

```python
import random

def multiplication_quiz():
    # Generate two random numbers between 1 and 10
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)

    # Ask the user to input the product of the two numbers
    print(f"What is the product of {num1} and {num2}?")
    user_answer = int(input("Your answer: "))

    # Calculate the correct product
    correct_answer = num1 * num2

    # Check if the user's answer is correct
    if user_answer == correct_answer:
        print("Correct! Well done.")
    else:
        print(f"Incorrect. The correct answer is {correct_answer}.")

# Run the quiz
multiplication_quiz()

```

Exercise 3: Squares of Even/Odd Numbers

```python
for num in range(100, 201):
    if num % 2 == 0:  # Check if the number is even
        print(f"The square of {num} is {num ** 2}")

```

Exercise 4: Word counter

```python
def count_words(text):
    # Split the text by spaces to get individual words
    words = text.split()

    # Return the number of words
    return len(words)

# Input from the user
user_text = input("Enter some text: ")

# Count the number of words and display the result
word_count = count_words(user_text)
print(f"The number of words in the text is: {word_count}")
```

Exercise 5: Check for Palindrome

```python
import string

def is_palindrome(s):
    # Convert the string to lowercase and remove spaces, punctuation, etc.
    s = s.lower()
    s = ''.join(char for char in s if char.isalnum())

    # Check if the string reads the same forward and backward
    return s == s[::-1]

# Input from the user
user_input = input("Enter a word or phrase: ")

# Check if it's a palindrome
if is_palindrome(user_input):
    print("The input is a palindrome.")
else:
    print("The input is not a palindrome.")

```
